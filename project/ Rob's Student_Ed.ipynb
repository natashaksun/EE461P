```
from sklearn.impute import SimpleImputer
import numpy as np
impute_values = [np.nan, -4, -5, -6, -7, -8, -9 ]

def run_all_imputers(column):
  for val in impute_values:
    imputer = SimpleImputer(missing_values=val, strategy='median')
    column = imputer.fit_transform(column)
  return column

def check_column(column):
  for val in impute_values:
    if val in column:
      print('Found {0} in column!!'.format(val))
      
def remove_missing_targets(df, targets):
  for target in targets:
    for val in impute_values:
      df = df[df[target] != val]
  return df

def replace_missing_values(df):
  df.replace(impute_values, value=np.nan, inplace=True)
```

```
import pandas as pd
#from sklearn.preprocessing import Imputer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

def impute_column(df, feature, threshold=10, enable_scaling=False):
    column = df[[feature]]
    unique_array = df[feature].unique()
    valid_uniques = [x for x in unique_array if x >= 0]
    if len(unique_array) > threshold or len(valid_uniques) <= 2:
        print('Imputing {0}'.format(feature))
        imputed_column = run_all_imputers(column)
        check_column(imputed_column)
        if enable_scaling and len(valid_uniques) > 2:
            print('Scaling {0}'.format(feature))
            scalar = StandardScaler()
            imputed_column = scalar.fit_transform(imputed_column)
        df.loc[:, feature] = imputed_column
        return df
    else:  # encode
        cat = [[x for x in unique_array if x >= 0]]
        cat = [sorted(cat[0])]
        if len(cat[0]) <= 2:
            return df
        print("Encoding {0}".format(feature))
        new_df = pd.get_dummies(df, prefix=feature, columns=[feature], drop_first=False)
        df.drop(feature, axis=1, inplace=True)
        for column in new_df.columns:
            df.loc[:, column] = new_df[column]
        return df
      
def impute_and_encode(df, targets, enable_scaling=False):
  for column in list(df):
    if column in targets:
      continue
    df = impute_column(df, column, threshold=15, enable_scaling=enable_scaling)
  return df
      
#test_df = pd.read_csv('/content/drive/Team Drives/EE461P/hsls_16_student_v1_0.csv')
targets = ['X3TGPAENG',
  'X3TGPAMAT',
  'X3TGPASCI',
  'X3TGPAHISCI',
  'X3TGPASOCST',
  'X3TGPAART',
  'X3TGPALANG',
  'X3TGPAACAD',
  'X3TAFGPATOT',
  'X4HSCOMPSTAT',
  'X4EVRAPPCLG',
  'X4PS1SELECT',
  'X4PS1DEGTYPE1',
  'X4PSLFSTFB16'
           ]
#targets=['X3TGPAACAD', 'X4PS1SELECT']
#X_imputed = student_no_null_or_weight
print(X_imputed.shape)
#X_imputed = remove_missing_targets(X_imputed, targets)
replace_missing_values(X_imputed)
print(X_imputed.shape)
X_imputed = impute_and_encode(X_imputed, targets, enable_scaling=True)
print(X_imputed.shape)
X_imputed.to_csv('/content/drive/Team Drives/EE461P/Imputed_scaled_all.csv')
```

```
def get_best_n_features(df, model, n=10):
  best_features = {}
  for i in range(n):
    feat_importance = model.feature_importances_.tolist()[i]
    larger_count = 0
    want_feature = True
    for val in model.feature_importances_:
      if feat_importance < val:
        larger_count += 1
      if larger_count > n:
        want_feature = False
        break
    if want_feature:
      best_features[str(larger_count)] = df.columns[i]
  return best_features

def get_best_features(X_class_train, model, n_features=10): 
  feature_importance = pd.DataFrame({'Features': X_class_train.columns, 'Coefficients': np.array(model.feature_importances_).astype(np.float)})
  feature_importance.sort_values(by=['Coefficients'], ascending = False, inplace = True)
  feature_importance.drop(feature_importance.tail(len(feature_importance.index)-n_features).index, inplace=True)
    
  display(feature_importance)
  return feature_importance

def get_best_reg_features(X_class_train, model, n_features=10): 
  feature_importance = pd.DataFrame({'Features': X_class_train.columns, 'Coefficients': np.array(model.coef_).astype(np.float)})
  feature_importance.sort_values(by=['Coefficients'], ascending = True, inplace = True)
  feature_importance.drop(feature_importance.tail(len(feature_importance.index)-n_features).index, inplace=True)
    
  display(feature_importance)
  return feature_importance
        
```

```
from sklearn.metrics import accuracy_score

def do_model_test(model, X_train, X_test, y_train, y_test):
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  return accuracy_score(y_test, y_pred)
  
```
